```python
import sys
import numpy as np
import pandas as pd
import matplotlib.image as mpimg

from IPython.display import display,HTML
```

```python
%matplotlib inline 
import matplotlib as plt
plt.rcParams["figure.figsize"] = (20,10)
```

```python
pkg_path = '/home/xgboost/workspace/xgb-arrow/xgboost/python-package'
sys.path.insert(0, pkg_path)
import xgboost as xgb
```

## XGBoost 1.3.3 model

```python
bst = xgb.Booster(model_file='/home/xgboost/models/xgboost-1.3.3-mortgage.model')
tree_df = bst.trees_to_dataframe()

tree_df_nodes = tree_df[['Tree', 'Node', 'Feature']]
tree_df_nodes = pd.DataFrame({'node_count': tree_df_nodes.groupby('Tree')['Node'].count(), 'leaf_count': tree_df_nodes.loc[tree_df_nodes['Feature'] == 'Leaf'].groupby('Tree')['Node'].count()})
tree_df_nodes.plot(kind='bar', figsize=(30,8))
```

```python
xgb.plot_importance(bst)
```
